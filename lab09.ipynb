```python
df_final=pd.get_dummies(df,['purpose'],drop_first=True)
```

```python
df.shape, df_final.shape, df_final.columns[14:]
```

#Train and test split

```python
from sklearn.model_selection import train_test_split
X=df_final.drop('not.fully.paid',axis=1)
y=df_final['not.fully.paid']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=150)
```

```python
df.shape,X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier(criterion='gini',max_depth=None)
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions=dtree.predict(X_test)
train_pred=dtree.predict(X_train)
```

```python
from sklearn.metrics  import classification_report,confusion_matrix,confusion_matrix,accuracy_score
```

```python
print(f"Accuracy of Train:{accuracy_score(y_train,train_pred)}")
print(f"Accuracy of Test:{accuracy_score(y_test,predictions)}")
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=600)
```

```python
rfc_pred=rfc.predict(X_test)
rfc_train_pred=rfc.predict(X_train)
```

```python
print(f"Accuracy of Tain:{accuracy_score(y_train,rfc_train_pred)}")
print(f"Accuracy of Tese:{accuracy_score(y_test,rfc_pred)}")
```

```python
cr=classification_report(y_test,rfc_pred)
print(f"Missclassification Error:{np.mean(rfc_pred!=y_test).sum()}")
```
