Let's suppose we collected age (in years), height (in cm), and weight (in kilograms) information from few patients visitng a clinic.

```python
people =['sai','rohan','nithin','shashank','shushanth','manideep','maneesh','varshith']
age = [13,21,18,41,32,99,142,56]
weight = [32,43,65,76,89,95,154,66]
height=[145,159,121,135,123,98,79,143]
```

```python
plt.scatter(weight , height)
plt.xlabel('weight')
plt.ylabel('height')
plt.title('weight vs height')
plt.show()

```

```python
plt.figure(figsize=(9,5))
plt.bar(people, weight)
plt.xlabel('people')
plt.ylabel('weight')
plt.title('people vs weight')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.hist(age , alpha=0.75 , color = 'red', bins=3)
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Age vs Height')
plt.show()

```
